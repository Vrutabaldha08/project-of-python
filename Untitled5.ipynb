# Convert km into miles 

```python
#collection of km
km = float(input("enter kilometre: "))

#conversion factor
conv_factor = 0.621371

#calculate in miles
miles = km * conv_factor

#print km in miles
print('%0.3f km is equal %0.3f miles' %(km,miles))
```

# Convert miles into km

```python
#collection of miles
miles = float(input("enter miles: "))

#conversion factor
conv_factor = 0.621371

#calculate in km
km = miles / conv_factor

#print miles in km
print('%0.3f miles is equal %0.3f km' %(miles,km))
```
